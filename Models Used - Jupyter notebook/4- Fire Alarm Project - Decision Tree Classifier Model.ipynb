```python
## a function to detect outliers
def find_outliers_IQR(df):
   q1=df.quantile(0.25)
   q3=df.quantile(0.75)
   IQR=q3-q1
   outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
   return outliers
```

```python
find_outliers_IQR(df['PM1.0']).count()
```

```python
## A library to detect outliers
!pip install datasist
```

```python
## we will change all the outlier values of all the columns in one step to be the median value
```

```python
from datasist.structdata import detect_outliers
```

```python
for i in df.columns:
    outlier_indices= detect_outliers(df,0,[i])
    df[i].iloc[outlier_indices]= df[i].median()
```

# Steps of machine learning after we finished the data analysis & cleaning the data:

## 1) Spliting the data to features & target

```python
features= df.drop('Fire Alarm', axis=1)
target= df['Fire Alarm']
```

## 2) Splitting the features & target to test & train

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(features, target, test_size=0.2, shuffle= True, stratify= target)
```

```python
## No need to down sampling or over sampling
```

## 3) Feature Scaling:
### we will apply normalization method because we already dealed with the outliers so there will be no need to be afraid from the effection of the outliers on the feature scaling during the use of normalization method.

```python
from sklearn.preprocessing import MinMaxScaler
scaling= MinMaxScaler()
x_train_scaled= scaling.fit_transform(x_train)
x_test_scaled= scaling.fit_transform(x_test)
```

# 4) Model selection:
## we'll use DesicionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
model= DecisionTreeClassifier(max_depth= 7)
model.fit(x_train_scaled, y_train)
```

```python
y_pred= model.predict(x_test_scaled)
print(np.sum(y_pred))
print(np.sum(y_test))
```

# 5) Accuracy

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
print('Accuracy= ', r2_score(y_test, y_pred))
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```
